```python
# Fill numeric columns using median
df["training_hours"].fillna(df["training_hours"].median(), inplace=True)
```

```python
df["satisfaction_score"].fillna(df["satisfaction_score"].median(), inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["overtime_hours"] = np.where(df["overtime_hours"] > 25, 25, df["overtime_hours"])
```

```python
df.isnull().sum()
df.describe()
```

```python
df.to_csv("../data/cleaned_data.csv", index=False)
```

```python
# Check attrition rate
attrition_rate = df["attrition"].mean()*100
print(f"Overall Attrition Rate: {attrition_rate:.2f}%")
```

```python
# Average satisfaction per department
avg_satisfaction = df.groupby("department")["satisfaction_score"].mean()
print(avg_satisfaction)
```
